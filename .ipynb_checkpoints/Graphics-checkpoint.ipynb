### Análise das variáveis

<b> Função

```python
def plot_variaveis_binarias(df_log_high, 
                            df_log_medium, 
                            df_log_low, 
                            variaveis, 
                            nome_imagem,
                            atributo_label,
                            target,
                            fig_width,
                            fig_height):
    """
    Gera um DataFrame de contagens (Target=1/0) para as variáveis binárias
    em três bases (High, Medium, Low), e plota um gráfico de barras horizontais 
    empilhadas (representando proporção Target=1/0) para cada variável e base.

    Parâmetros:
    -----------
    df_hiring_log_high: pd.DataFrame
        DataFrame da base "High".
    df_hiring_log_medium: pd.DataFrame
        DataFrame da base "Medium".
    df_hiring_log_low: pd.DataFrame
        DataFrame da base "Low".
    variaveis: list
        Lista de strings com o nome das variáveis binárias que serão analisadas.
    nome_imagem: str
        Nome (caminho) para salvar a imagem do gráfico. Ex.: "grafico.png"

    Retorno:
    --------
    None
    (A função exibe o gráfico e salva a figura no arquivo especificado.)
    """
    # -----------------------------
    # Passo 1: Gerar as contagens
    # -----------------------------

    target_col = target  # Nome da coluna binária de resultado
    linhas_resultado = []

    # Para cada variável, calcula quantos registros possuem:
    # - var=1 e Target=1
    # - var=1 e Target=0
    # - var=0 e Target=1
    # - var=0 e Target=0
    for var in variaveis:
        linha = {'Variavel': var}

        for nome_df, df in [('High', df_log_high),
                            ('Medium', df_log_medium),
                            ('Low', df_log_low)]:

            count_1_1 = len(df[(df[var] == 1) & (df[target_col] == 1)])
            count_1_0 = len(df[(df[var] == 1) & (df[target_col] == 0)])
            count_0_1 = len(df[(df[var] == 0) & (df[target_col] == 1)])
            count_0_0 = len(df[(df[var] == 0) & (df[target_col] == 0)])

            linha[f'{nome_df}_1_1'] = count_1_1
            linha[f'{nome_df}_1_0'] = count_1_0
            linha[f'{nome_df}_0_1'] = count_0_1
            linha[f'{nome_df}_0_0'] = count_0_0

        linhas_resultado.append(linha)

    df_resultado = pd.DataFrame(linhas_resultado)

    # -----------------------------
    # Passo 2: Organizar dados para plot
    # -----------------------------
    # Cada variável terá 6 linhas no eixo Y:
    #   High(1), High(0), Medium(1), Medium(0), Low(1), Low(0)
    base_varcat_pairs = [
        ('High', 1),
        ('High', 0),
        ('Medium', 1),
        ('Medium', 0),
        ('Low', 1),
        ('Low', 0),
    ]

    linhas_plot = []
    for _, row in df_resultado.iterrows():
        var_name = row['Variavel']

        sub_idx = 0  # posição dentro de 6 linhas para cada variável
        for (base, varcat) in base_varcat_pairs:
            # Colunas de contagem Target=1 / Target=0
            col_t1 = f'{base}_{varcat}_1'
            col_t0 = f'{base}_{varcat}_0'

            count_t1 = row[col_t1]
            count_t0 = row[col_t0]
            total = count_t1 + count_t0

            # Cálculo de proporções (empilhamento de 0..1 no eixo X)
            if total > 0:
                prop_t1 = count_t1 / total
                prop_t0 = count_t0 / total
            else:
                prop_t1 = 0
                prop_t0 = 0

            # Construção da label no eixo Y
            if sub_idx == 0:
                label_y = f"{var_name:<50}{base} {varcat}" #Ajusta o espaço para o nome da variável
            elif varcat == 1:
                label_y = f"{' ':<25}{base} {varcat}"
            else:
                label_y = f"{' ':<25}{' '*(len(base)+1)}{varcat}"

            linhas_plot.append({
                'variavel': var_name,
                'base': base,
                'varcat': varcat,
                'sub_idx': sub_idx,
                'count_t1': count_t1,
                'count_t0': count_t0,
                'prop_t1': prop_t1,
                'prop_t0': prop_t0,
                'label_y': label_y
            })

            sub_idx += 1

    # -----------------------------
    # Passo 3: Plot das barras horizontais empilhadas
    # -----------------------------
    y_positions = range(len(linhas_plot))
    labels_y = [d['label_y'] for d in linhas_plot]

    prop_t1_list = [d['prop_t1'] for d in linhas_plot]
    prop_t0_list = [d['prop_t0'] for d in linhas_plot]

    plt.figure(figsize=(fig_width, fig_height))

    # Duas cores (tons de lilás/roxo)
    colors = ['#9999ea', '#d4d4fa']

    # Barras de Target=1
    plt.barh(y_positions, prop_t1_list, color=colors[0], label='Positive')
    # Barras de Target=0 (empilhadas)
    plt.barh(y_positions, prop_t0_list, left=prop_t1_list, color=colors[1], label='Negative')

    plt.yticks(ticks=list(y_positions), labels=labels_y)
    plt.xlabel('Proportion')
    plt.ylabel(f"{atributo_label} | Event log | Label")
    # Removendo o título, ou defina se quiser: plt.title('Meu Gráfico')

    # Construindo manualmente a legenda para ter "Case Result" em linha
    # 1) Um "Patch" invisível para servir de "título"
    title_patch = mpatches.Patch(color='none', label='Case outcome:')
    # 2) Patches para as duas categorias
    positive_patch = mpatches.Patch(color=colors[0], label='Positive')
    negative_patch = mpatches.Patch(color=colors[1], label='Negative')

    # 3) Montando a legenda com 3 itens em 1 linha
    plt.legend(handles=[title_patch, positive_patch, negative_patch],
               loc='lower center',
               bbox_to_anchor=(0.5, -0.05),
               ncol=3)

    # Inverte o eixo Y para que a primeira linha apareça no topo
    plt.gca().invert_yaxis()

    # Desenhar linhas horizontais de separação entre bases e variáveis
    ax = plt.gca()
    num_barras = len(linhas_plot)

    for i, item in enumerate(linhas_plot):
        sub_i = item['sub_idx']
        var = item['variavel']

        # Se não for a última linha
        if i < len(linhas_plot) - 1:
            var_next = linhas_plot[i+1]['variavel']
            # Se mudamos de variável no sub_idx=5 => linha forte
            if var != var_next and sub_i == 5:
                ax.axhline(i + 0.5, color='black', linewidth=1.5)
            else:
                # Se ainda estamos na mesma variável e sub_i em [1,3], linha fina
                if sub_i in [1, 3]:
                    ax.axhline(i + 0.5, color='gray', linewidth=1)
        # Se for a última linha, não há próxima

    # Ajuste o ylim para "colar" as barras no topo e na base
    ax.set_ylim([num_barras - 0.5, -0.5])

    # Ajusta a plotagem e salva a figura
    plt.tight_layout()
    plt.savefig(nome_imagem, dpi=300)  # Salva com resolução de 300 dpi
    plt.show()
```

<b> Análise das transições

```python
def plot_transicoes_binarias(df_log_high, 
                            df_log_medium, 
                            df_log_low, 
                            variaveis, 
                            nome_imagem,
                            atributo_label,
                            target,
                            fig_width,
                            fig_height):
    """
    Gera um gráfico de barras horizontais empilhadas para as variáveis binárias
    nas três bases (High, Medium, Low), considerando apenas os registros onde
    a transição ocorreu (variável = 1). O eixo Y mantém o formato:

        'Atividade A -> Atividade B'  High
                                      Medium
                                      Low

    Parâmetros:
    -----------
    df_log_high: pd.DataFrame
        DataFrame da base "High".
    df_log_medium: pd.DataFrame
        DataFrame da base "Medium".
    df_log_low: pd.DataFrame
        DataFrame da base "Low".
    variaveis: list
        Lista de variáveis binárias a serem analisadas.
    nome_imagem: str
        Nome do arquivo para salvar o gráfico.
    atributo_label: str
        Rótulo para o eixo Y.
    target: str
        Nome da variável binária que representa o resultado do case.
    fig_width: float
        Largura da figura.
    fig_height: float
        Altura da figura.

    Retorno:
    --------
    None
    (A função exibe o gráfico e salva a imagem.)
    """

    target_col = target  
    linhas_resultado = []

    for var in variaveis:
        linha = {'Variavel': var}

        for nome_df, df in [('High', df_log_high),
                            ('Medium', df_log_medium),
                            ('Low', df_log_low)]:

            # Considera apenas os registros onde a variável = 1 (transição ocorreu)
            df_filtered = df[df[var] == 1]

            count_t1 = len(df_filtered[df_filtered[target_col] == 1])  # Target positivo
            count_t0 = len(df_filtered[df_filtered[target_col] == 0])  # Target negativo

            linha[f'{nome_df}_1'] = count_t1
            linha[f'{nome_df}_0'] = count_t0

        linhas_resultado.append(linha)

    df_resultado = pd.DataFrame(linhas_resultado)

    # Organizando os dados para o gráfico
    linhas_plot = []
    for _, row in df_resultado.iterrows():
        var_name = row['Variavel']

        for base in ['High', 'Medium', 'Low']:
            count_t1 = row[f'{base}_1']
            count_t0 = row[f'{base}_0']
            total = count_t1 + count_t0

            prop_t1 = count_t1 / total if total > 0 else 0
            prop_t0 = count_t0 / total if total > 0 else 0

            # Criando o formato correto para o eixo Y
            if base == "High":
                label_y = f"{var_name:<60}  {base}"  # High aparece na mesma linha da transição
            else:
                label_y = f"  {base}"  # Medium e Low aparecem abaixo, alinhados

            linhas_plot.append({
                'variavel': var_name,
                'base': base,
                'count_t1': count_t1,
                'count_t0': count_t0,
                'prop_t1': prop_t1,
                'prop_t0': prop_t0,
                'label_y': label_y
            })

    # Criando o gráfico de barras horizontais empilhadas
    y_positions = range(len(linhas_plot))
    labels_y = [d['label_y'] for d in linhas_plot]

    prop_t1_list = [d['prop_t1'] for d in linhas_plot]
    prop_t0_list = [d['prop_t0'] for d in linhas_plot]

    plt.figure(figsize=(fig_width, fig_height))

    colors = ['#9999ea', '#d4d4fa']  # Tons de lilás/roxo

    plt.barh(y_positions, prop_t1_list, color=colors[0], label='Positive')
    plt.barh(y_positions, prop_t0_list, left=prop_t1_list, color=colors[1], label='Negative')

    plt.yticks(ticks=list(y_positions), labels=labels_y)
    plt.xlabel('Proportion')
    plt.ylabel(f"{atributo_label} | Event log")

    # Criando legenda personalizada
    title_patch = mpatches.Patch(color='none', label='Protected Attribute')
    positive_patch = mpatches.Patch(color=colors[0], label='Protected group')
    negative_patch = mpatches.Patch(color=colors[1], label='Unprotected group')

    plt.legend(handles=[title_patch, positive_patch, negative_patch],
               loc='lower center',
               bbox_to_anchor=(0.5, -0.05),
               ncol=3)

    plt.gca().invert_yaxis()

    # Adicionando linhas de separação entre diferentes transições
    ax = plt.gca()
    num_barras = len(linhas_plot)

    for i, item in enumerate(linhas_plot[:-1]):
        if item['variavel'] != linhas_plot[i + 1]['variavel']:
            ax.axhline(i + 0.5, color='black', linewidth=1.5)

    ax.set_ylim([num_barras - 0.5, -0.5])

    plt.tight_layout()
    plt.savefig(nome_imagem, dpi=300)
    plt.show()
```

```python
def plot_transicoes_binarias(df_log_high, 
                             df_log_medium, 
                             df_log_low, 
                             variaveis, 
                             nome_imagem,
                             atributo_label,
                             target,
                             fig_width,
                             fig_height,
                             protected_original):
    """
    Gera um gráfico de barras horizontais empilhadas para as variáveis binárias
    nas três bases (High, Medium, Low), considerando apenas os registros onde
    a transição ocorreu (variável = 1), e adiciona uma linha vertical
    tracejada vermelha em cada barra, na posição definida em protected_original.

    Parâmetros adicionais:
    ----------------------
    protected_original: dict
        Um dicionário mapeando cada base ('High', 'Medium', 'Low') para um
        valor float entre 0 e 1, ex:
            {'High': 0.10, 'Medium': 0.20, 'Low': 0.30}
    """
    # --- cálculo das proporções (mantido igual) ---
    target_col = target  
    linhas_resultado = []
    for var in variaveis:
        linha = {'Variavel': var}
        for nome_df, df in [('High', df_log_high),
                            ('Medium', df_log_medium),
                            ('Low', df_log_low)]:
            df_f = df[df[var] == 1]
            linha[f'{nome_df}_1'] = len(df_f[df_f[target_col] == 1])
            linha[f'{nome_df}_0'] = len(df_f[df_f[target_col] == 0])
        linhas_resultado.append(linha)
    df_resultado = pd.DataFrame(linhas_resultado)

    linhas_plot = []
    for _, row in df_resultado.iterrows():
        var_name = row['Variavel']
        for base in ['High', 'Medium', 'Low']:
            t1 = row[f'{base}_1']; t0 = row[f'{base}_0']
            total = t1 + t0
            prop_t1 = t1/total if total>0 else 0
            prop_t0 = t0/total if total>0 else 0
            label = f"{var_name:<60}  {base}" if base=='High' else f"  {base}"
            linhas_plot.append({
                'variavel': var_name,
                'base': base,
                'prop_t1': prop_t1,
                'prop_t0': prop_t0,
                'label_y': label
            })

    # --- plotagem ---
    y_pos = range(len(linhas_plot))
    props1 = [d['prop_t1'] for d in linhas_plot]
    props0 = [d['prop_t0'] for d in linhas_plot]
    labels = [d['label_y'] for d in linhas_plot]

    fig, ax = plt.subplots(figsize=(fig_width, fig_height))
    ax.barh(y_pos, props1,   color='#9999ea', label='Positive')
    ax.barh(y_pos, props0,   left=props1, color='#d4d4fa', label='Negative')
    ax.set_yticks(y_pos);  ax.set_yticklabels(labels)
    ax.set_xlabel('Proportion')
    ax.set_ylabel(f"{atributo_label} | Event log")

    # legenda
    title_patch    = mpatches.Patch(color='none',    label='Protected Attribute')
    positive_patch = mpatches.Patch(color='#9999ea', label='Protected group')
    negative_patch = mpatches.Patch(color='#d4d4fa', label='Unprotected group')
    ax.legend(handles=[title_patch, positive_patch, negative_patch],
              loc='lower center', bbox_to_anchor=(0.5, -0.05), ncol=3)

    ax.invert_yaxis()

    # linhas de separação entre variáveis
    for i in range(len(linhas_plot)-1):
        if linhas_plot[i]['variavel'] != linhas_plot[i+1]['variavel']:
            ax.axhline(i+0.5, color='black', linewidth=1.5)

    # --- aqui: linhas tracejadas vermelhas em cada barra ---
    half_height = 0.4  # metade da espessura da barra (padrão barh)
    for i, item in enumerate(linhas_plot):
        base = item['base']
        x = protected_original.get(base, None)
        if x is not None:
            ax.vlines(x,
                      i - half_height,
                      i + half_height,
                      colors='red',
                      linestyles='--',
                      linewidth=1)

    ax.set_ylim([len(linhas_plot)-0.5, -0.5])
    plt.tight_layout()
    plt.savefig(nome_imagem, dpi=300)
    plt.show()
```

```python
def plot_transicoes_binarias(df_log_high, 
                             df_log_medium, 
                             df_log_low, 
                             variaveis, 
                             nome_imagem,
                             atributo_label,
                             target,
                             fig_width,
                             fig_height,
                             protected_original):
    """
    Gera um gráfico de barras horizontais empilhadas para as variáveis binárias
    nas três bases (High, Medium, Low), considerando apenas os registros onde
    a transição ocorreu (variável = 1), adiciona uma linha vertical
    tracejada vermelha em cada barra na posição definida em protected_original,
    e inclui uma legenda para essa linha (quebrada em duas linhas).
    """
    # --- cálculo das proporções ---
    target_col = target  
    linhas_resultado = []
    for var in variaveis:
        linha = {'Variavel': var}
        for nome_df, df in [('High', df_log_high),
                            ('Medium', df_log_medium),
                            ('Low', df_log_low)]:
            df_f = df[df[var] == 1]
            linha[f'{nome_df}_1'] = len(df_f[df_f[target_col] == 1])
            linha[f'{nome_df}_0'] = len(df_f[df_f[target_col] == 0])
        linhas_resultado.append(linha)
    df_resultado = pd.DataFrame(linhas_resultado)

    linhas_plot = []
    for _, row in df_resultado.iterrows():
        var_name = row['Variavel']
        for base in ['High', 'Medium', 'Low']:
            t1 = row[f'{base}_1']; t0 = row[f'{base}_0']
            total = t1 + t0
            prop_t1 = t1/total if total>0 else 0
            prop_t0 = t0/total if total>0 else 0
            label = f"{var_name:<60}  {base}" if base=='High' else f"  {base}"
            linhas_plot.append({
                'variavel': var_name,
                'base': base,
                'prop_t1': prop_t1,
                'prop_t0': prop_t0,
                'label_y': label
            })

    # --- plotagem ---
    y_pos = range(len(linhas_plot))
    props1 = [d['prop_t1'] for d in linhas_plot]
    props0 = [d['prop_t0'] for d in linhas_plot]
    labels = [d['label_y'] for d in linhas_plot]

    fig, ax = plt.subplots(figsize=(fig_width, fig_height))
    ax.barh(y_pos, props1,   color='#9999ea', label='Protected group')
    ax.barh(y_pos, props0,   left=props1, color='#d4d4fa', label='Non-protected group')

    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels)
    ax.set_xlabel('Proportion')
    ax.set_ylabel(f"{atributo_label} | Event log")
    ax.invert_yaxis()

    # legendas das barras
    title_patch    = mpatches.Patch(color='none',    label='Protected Attribute')
    positive_patch = mpatches.Patch(color='#9999ea', label='Protected group')
    negative_patch = mpatches.Patch(color='#d4d4fa', label='Unprotected group')

    # legenda da linha tracejada vermelha (quebrada em duas linhas)
    red_line = Line2D([0], [0],
                      color='red',
                      lw=1,
                      linestyle='--',
                      label='Proportion of protected group\nin the complete event log')

    ax.legend(handles=[title_patch, positive_patch, negative_patch, red_line],
              loc='lower center',
              bbox_to_anchor=(0.2, -0.07),
              ncol=4)

    # linhas de separação entre variáveis
    for i in range(len(linhas_plot)-1):
        if linhas_plot[i]['variavel'] != linhas_plot[i+1]['variavel']:
            ax.axhline(i+0.5, color='black', linewidth=1.5)

    # linhas tracejadas vermelhas em cada barra
    half_height = 0.4
    for i, item in enumerate(linhas_plot):
        base = item['base']
        x = protected_original.get(base, None)
        if x is not None:
            ax.vlines(x,
                      i - half_height,
                      i + half_height,
                      colors='red',
                      linestyles='--',
                      linewidth=1)

    ax.set_ylim([len(linhas_plot)-0.5, -0.5])
    plt.tight_layout()
    plt.savefig(nome_imagem, dpi=300, bbox_inches='tight', pad_inches=0.1)
    plt.show()            
```

```python
def ActivityTransitionsOneHotEncoding(df, var_id, var_activity, var_timestamp):
    """
    Esta função realiza a codificação one-hot para transições de atividades em um processo. 
    Dado um DataFrame que contém identificadores de casos, atividades e timestamps, 
    a função identifica as transições de uma atividade para outra dentro de cada caso 
    e cria uma matriz de presença dessas transições para cada caso. Além disso, 
    a função retorna o total de cada transição no DataFrame original.

    Parâmetros:
    - df (pd.DataFrame): DataFrame contendo os dados do processo.
    - var_id (str): Nome da coluna que contém os identificadores dos casos (padrão: 'case:case').
    - var_activity (str): Nome da coluna que contém as atividades (padrão: 'activity').
    - timestamp_col (str): Nome da coluna que contém os timestamps (padrão: 'time:timestamp').

    Retorna:
    - pd.DataFrame: DataFrame com a matriz de presença de transições por caso.
    - pd.DataFrame: DataFrame com o total de cada transição no DataFrame original.
    """

    # Converter a coluna de timestamp para datetime
    df[var_timestamp] = pd.to_datetime(df[var_timestamp])

    # Ordenar por caso e timestamp
    df = df.sort_values(by=[var_id, var_timestamp])

    # Identificar a próxima atividade para cada evento
    df['next_activity'] = df.groupby(var_id)[var_activity].shift(-1)

    # Criar a coluna de transições
    df['transition'] = df[var_activity] + ' -> ' + df['next_activity']

    # Remover transições onde 'next_activity' é NaN
    df = df.dropna(subset=['next_activity'])
    
    # Excluir a última transição de cada caso
    #df['is_last'] = df.groupby(var_id).cumcount(ascending=False) == 0
    #df = df[~df['is_last']].copy()

    # Marcar a presença de transições por caso
    transition_presence = df.groupby([var_id, 'transition']).size().reset_index(name='count')
    transition_presence['count'] = 1  # Substituir a contagem por 1 para marcar a presença

    # Pivotar os dados para criar a matriz de presença de transições
    transition_pivot = transition_presence.pivot(index=var_id, columns='transition', values='count').fillna(0)

    # Resetar o índice para que var_id vire uma coluna
    transition_pivot = transition_pivot.reset_index()

    # Contar o total de cada transição no DataFrame original
    transition_total_counts = df['transition'].value_counts().reset_index()
    transition_total_counts.columns = ['transition', 'total_count']

    return transition_pivot, transition_total_counts
```

```python
# lista de tuplas (DataFrame, nome do processo)
bases = [
    (df_hiring_resultados, "Hiring"),
    (df_hospital_resultados, "Hospital"),
#    (df_lending_resultados, "Lending"),
#    (df_renting_resultados, "Renting"),
]

# métricas escalares e vetoriais
metrics_scalar = [
    "F1-Score",
    "Max Difference - Equalized Odds (TPR)",
    "Max Difference - Equalized Odds (FPR)",
]
metrics_vector = [
    "Equalized Odds (TPR)",
    "Equalized Odds (FPR)",
]

rows = []
for df, processo in bases:
    df_test = df[df["Source"] == "Test"]
    for (modelo, base), grp in df_test.groupby(["Type", "Dataset"]):
        rec = {
            "Processo": processo,
            "Base": base,
            "Modelo": modelo,
        }
        # valores escalares
        for m in metrics_scalar:
            vals = grp.loc[grp["Metric"] == m, "Value"]
            rec[m] = vals.iloc[0] if not vals.empty else None

        # valores vetoriais
        for m in metrics_vector:
            vals = grp.loc[grp["Metric"] == m, "Value"]
            if vals.empty:
                rec[f"{m}_case:protected_1"] = None
                rec[f"{m}_case:protected_0"] = None
            else:
                raw = vals.iloc[0]
                # se for string, converte com literal_eval
                if isinstance(raw, str):
                    # espera algo como "{0: 0.8, 1: 0.6}" ou similar
                    try:
                        parsed = ast.literal_eval(raw)
                    except Exception:
                        # caso seja o print de uma Series, transforma em dict manualmente
                        # exc: "0    0.8\n1    0.6\n"
                        lines = [l.strip() for l in raw.splitlines() if l and l[0].isdigit()]
                        parsed = {}
                        for ln in lines:
                            k, v = ln.split()
                            parsed[int(k)] = float(v)
                else:
                    parsed = raw  # já é dict ou Series

                # extrai protegido=1 e =0
                rec[f"{m}_case:protected_1"] = parsed.get(1, None)
                rec[f"{m}_case:protected_0"] = parsed.get(0, None)

        rows.append(rec)

# monta DataFrame final
df_final = pd.DataFrame(rows, columns=[
    "Processo", "Base", "Modelo",
    "F1-Score",
    "Max Difference - Equalized Odds (TPR)",
    "Equalized Odds (TPR)_case:protected_1",
    "Equalized Odds (TPR)_case:protected_0",
    "Max Difference - Equalized Odds (FPR)",
    "Equalized Odds (FPR)_case:protected_1",
    "Equalized Odds (FPR)_case:protected_0",
])

df_final.head()
```

```python
#df2 = df_resultados[(df_resultados['Modelo'] != 'Baseline')]
#df2 = df2[(df2['Modelo'] != 'Reference')]

df2 = df_resultados[(df_resultados['Modelo'] != 'Q&A reference')]

# Dicionários para marcadores e cores
markers_dict = {
    #"Q&A reference": "o",
    "No-mitigation reference": "s",
    "Pre-processing: Reweighing": "v",
    "Pre-processing: Disparate Impact Remover": "^",
    "In-processing: Adversarial Debiasing": "D",
    "Post-processing: Equalized Odds": "<",
    "Post-processing: Calibrated EqOdds": ">"
}

colors_dict = {
    #"Q&A reference": "#1d1d27",
    "No-mitigation reference": "#b3b3f0",
    "Pre-processing: Reweighing": "#333349",
    "Pre-processing: Disparate Impact Remover": "#4b4a6e",
    "In-processing: Adversarial Debiasing": "#646395",
    "Post-processing: Equalized Odds": "#7e7ebf",
    "Post-processing: Calibrated EqOdds": "#9999ea"
}

# ──────────────────────────────────────────────────────────────────────────────
# 1)  PREPARE TODOS OS HANDLES DA LEGENDA
# ──────────────────────────────────────────────────────────────────────────────

# handles para cada modelo (mesma cor e mesmo marcador usados no scatter)
model_handles = [
    Line2D([], [],                      # lista vazia = nada a ser plotado
           marker=markers_dict[m],      # marcador do modelo
           color='none',                # linha invisível
           markerfacecolor=colors_dict[m],
           markeredgecolor=colors_dict[m],
           markersize=7,
           linestyle='',                # sem linha
           label=m)                     # texto na legenda
    for m in markers_dict
]

# handle para a linha tracejada vermelha (baseline)
#baseline_handle = Line2D([], [], 
#                         color='red', 
#                         linestyle='--', 
#                         linewidth=1.5,
#                         label='Original data')  

# Junta todos os handles numa lista
#legend_handles = [baseline_handle] + model_handles

# ──────────────────────────────────────────────────────────────────────────────
# 2)  CRIE O FACETGRID E DESENHE
# ──────────────────────────────────────────────────────────────────────────────

g = sns.FacetGrid(
    data=df2,
    row="Processo",
    col="Base",
    margin_titles=True
)

def facet_scatter(data, **kwargs):
    # pontos
    sns.scatterplot(
        data=data,
        x="F1-Score",
        y="Equalized Odds (FPR)",
        hue="Modelo",
        style="Modelo",
        markers=markers_dict,
        palette=colors_dict,
        s=70,
        **kwargs
    )
    
g.map_dataframe(facet_scatter)

# ──────────────────────────────────────────────────────────────────────────────
# 3)  ADICIONE A LEGENDA COM TODOS OS HANDLES
# ──────────────────────────────────────────────────────────────────────────────

g.add_legend(
    handles=model_handles,             # ➌  inclui modelos + baseline
    title="Models",         # título geral da legenda
    ncol=3,
    loc="upper center",      
    bbox_to_anchor=(0.3, 0.15)
)

# Define limites dos eixos
g.set(xlim=(-0.1, 1.1), ylim=(-0.1, 1.1))

# Remove prefixos "Processo = " e "Base = " nos títulos de cada faceta
g.set_titles(row_template='{row_name}', col_template='{col_name}')

# Para ganhar espaço na parte de baixo (onde fica a legenda)
plt.subplots_adjust(bottom=0.2)

# Adiciona um título geral para o "eixo das colunas"
# (por exemplo, "Discrimination level" no centro, abaixo do grid)
g.fig.text(
    0.28,      # posição X no sistema de coordenadas da figura (0.5 = meio)
    1.0,     # posição Y (mais alto ou baixo dependendo do espaço disponível)
    'Discrimination level', 
    ha='center'
)

# Adiciona um título geral para o "eixo das linhas"
# (por exemplo, "Event log" na esquerda, em posição central vertical)
g.fig.text(
    0.5,     # posição X 
    0.6,      # posição Y (meio vertical)
    'Event log',
    va='center',
    rotation=-90
)

plt.savefig("fig_metric_EqualizedOdds_FPR.png", dpi=300, bbox_inches="tight")
plt.show()
```
